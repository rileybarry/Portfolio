```python
nba_series = pd.read_csv('nba.csv')
nba_series = nba_series.set_index('Name')
nba_series = nba_series.drop(columns=['Team','Number','Position','Age','Height','Weight','College'])
nba_series.head()
```

```python
#2c
import math

def sal_func(sal):
    av = nba_series['Salary'].mean()
    if math.isnan(sal) == True:
        return 'No Salary Reported'    
    elif sal == av:
        return 'Average salary'
    elif sal < av:
        return 'Below average'
    elif sal > av:
        return 'Above average'
    
#nba_series['Average?'] = sal_func(new_nobel['Salary'])
nba_series['Average?'] = nba_series['Salary'].apply(sal_func)
nba_series.head()
```

```python
#4a

new_nba = pd.read_csv('nba.csv')
new_nba = new_nba.sample(frac=0.8)
new_nba.info()
```

```python
#4e

new_nba['Salary']=new_nba['Salary'].astype(int)
new_nba['Number']=new_nba['Number'].astype(int)
new_nba.info()
```
