## Question 1:
- Create  Numpy arrays for, x,y, where x and y are equal and both of them are between [0,11], and z = x^2 

### 1a) Create the first plot by creating a figure object located at [0,0,1,1] and plot (x,y) on this location and set the labels of your choice and make the title " x- y relation".  (as shown below!)




```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(0, 11)
y = x
z = x ** 2

f = plt.figure()
axis = f.add_axes([0, 0, 1, 1])

axis.plot(x, y)
axis.set_xlabel('x')
axis.set_ylabel('y')
axis.set_title('x-y relation')
```


### 1b) Now, create a figure object and put two axes on it, ax1 and ax2, which are located at [0,0,1,1] and [0.3,0.6,.3,.3] respectively. Now, plot (x,z) on the outer axes (Located at [0,0,1,1]) and plot (x,y) on the inner axes (located at [0.3,0.6,.3,.3]). Set the labels of your choice. Make your titles as "x- z relation" for the outer axes and "x- y relation" for the inner axes.  (as shown below!)

```python
f = plt.figure()
ax1 = f.add_axes([0, 0, 1, 1])
ax2 = f.add_axes([0.3, 0.6, .3, .3])

ax1.plot(x, z)
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax1.set_title('x-z relation')

ax2.plot(x, y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('x-y relation')
```

## Question 2a)

### 2a) Use plt.subplots(nrows=1, ncols=2) to plot ((x,y) and (x,z) (as shown below!)
- Note that x=y, and z = x ** 2  (same x, y, z values that we used for question 1a, 1b, and 2a)
- (Note that the Figure size is (12,6) and,  line width (for both figures are 3)


```python
f, axis = plt.subplots(figsize = (12, 6), nrows = 1, ncols = 2)
axis[0].plot(x, y,
            'b',
            lw = 3)
axis[0].set_xlabel('x')
axis[0].set_ylabel('y')
axis[0].set_ybound(lower = 0, upper = 10)

axis[1].plot(x, z,
            'r--',
            lw = 3)
axis[1].set_xlabel('x')
axis[1].set_ylabel('z')

plt.tight_layout()
```

## Question 2b) 

### 2b) Create the figure below by creating a for loop and using .subplots function
- Note that x=y, and z = x ** 2  (same x, y, z values that we used for question 1a, 1b, and 2a)

```python
f, axis = plt.subplots(nrows=1, ncols=2)

for item in axis:
    item.plot(x,z,label='X-Z Relation')
    item.plot(x,y,label='X-Y Relation')
    item.legend(loc=1)
```
